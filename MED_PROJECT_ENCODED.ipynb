## here i will encode categorical columns according to the following:
#### 1-company, operation system and laptop type(TypeName) will be encoded by binary encoder since their quality and performance are debatable
#### 2-the rest of categoricals will be encoded by mapping

### company encoding

```python
encoder=be()
```

```python
company_enc=encoder.fit_transform(df[["Company"]])
```

```python
df=pd.concat([df,company_enc],axis=1)
```

```python
df.drop("Company",axis=1,inplace=True)
```

```python
df
```

### type name encoding

```python
type_enc=encoder.fit_transform(df[["TypeName"]])
```

```python
df=pd.concat([df,type_enc],axis=1)
```

```python
df.drop("TypeName",axis=1,inplace=True)
```

```python
OS_enc=encoder.fit_transform(df[["OpSys"]])
```
